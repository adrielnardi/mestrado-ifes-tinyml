```python
#USO DAS BATERIAS COM SENSOR CAMERA

import pandas as pd
import matplotlib.pyplot as plt
# Create a list of data
# to be represented in x-axis
hrs = [ 0,1,2,3,4,5,6 ]

# Create a list of data
# to be represented in y-axis
bat1 = [3.98,3.95,3.89,3.82,3.7,3.69,2.16]
bat2 = [4.06,4.00,3.96,3.89,3.7,3.67,1.98]
bat3 = [3.77,3.71,3.35,0.60,0.35,0.35,0.17]


# Create a dataframe using the two lists
df_days_calories = pd.DataFrame(
    { 'Horas (h)' : hrs ,
     'Bateria 1': bat1 ,
     'Bateria 2': bat2 ,
     'Bateria 3': bat3})

ax = plt.gca()

#use plot() method on the dataframe
df_days_calories.plot( x = 'Horas (h)'  , y = 'Bateria 1', ax = ax )
df_days_calories.plot( x = 'Horas (h)'  , y = 'Bateria 2' , ax = ax )
df_days_calories.plot( x = 'Horas (h)'  , y = 'Bateria 3' , ax = ax )


## CICLO C/ CAMERA
# 11.81 v total que começou
# 4.31 v total que terminou

#7.5v gastos durante 6 horas
#1.25 volts/hora de consumo
```
